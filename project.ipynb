```python
import os
from PIL import Image
from transformers import ViTImageProcessor, ViTForImageClassification

#путь к основной папке с датасетом PlantVillage(должна лежать рядом с файлом projectipynb)
root_folder = "PlantVillage"

#скачиваем и подгружаем модель
model_name = "Carina124/plant-vit-model-3"
processor = ViTImageProcessor.from_pretrained(model_name)
model = ViTForImageClassification.from_pretrained(model_name)

count = 0
max_images = 15

done = False

#рекурсивно обходим подпапки внутри датасета
for subdir, _, files in os.walk(root_folder):
    if done:
        break
    for file in files:
        if file.lower().endswith(('.jpg')):
            img_path = os.path.join(subdir, file)
            
            #открываем изображение и конвертируемт только в red green blue поток, тем самым убирая прозрачность 
            image = Image.open(img_path).convert("RGB")
       
            #применяем предобработку для возвращаем тензор pt=pytorch
            inputs = processor(images=image, return_tensors="pt")
               
            #передаём тензор в модель и возвращаем атрибут logist
            logits = model(**inputs).logits
           
            #находим индекс класса с максимальным logits и превращаем его в обычное число
            idx = logits.argmax(-1).item()
        
            #получаем ответ
            label = model.config.id2label.get(idx, "Unknown")

            #выводим
            print(f"Файл картинки: {img_path} \nПредсказание: {label} \n")
            
            #закоментировать этот блок кода для проверки по всем файлам, либо увеличить число max_images
            count += 1
            if count >= max_images:
                done = True
                break
```
